```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px 

```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle='',marker='s',markersize=10,alpha=.5,figsize=(5,5),markerfacecolor='purple',markeredgecolor='black')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(10,6),color='yellow')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='area',figsize=(15,4),color='orange',alpha=0.5)
```

```python
print(sns.get_dataset_names())
```

```python
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5)
```

# Subplot

```python
fig,axs=plt.subplots(1,2,figsize=(15,4),constrained_layout=True)
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5,ax=axs[0],color='red',title='penguin bill data')
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5,ax=axs[1],color='black',title='penguins flipper')
plt.show()
```
