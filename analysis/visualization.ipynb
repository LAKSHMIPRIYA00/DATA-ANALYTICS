# Subplot

```python
fig,axs=plt.subplots(1,2,figsize=(15,4),constrained_layout=True)
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5,ax=axs[0],color='red',title='penguin bill data')
penguins.plot(kind='scatter',x='bill_length_mm',y='flipper_length_mm',s=100,alpha=.5,ax=axs[1],color='black',title='penguins flipper')
plt.show()
```
